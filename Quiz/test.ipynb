Câu 4 : Ý tưởng của Dynamic Programming? Multi select

A.	Là một kĩ thuật thiết kế thuật toán theo kiểu chia bài toán lớn thành các bài toán con, sử dụng lời giải của các bài toán con để tìm lời giải cho bài toán ban đầu.

B.	Là sự cải tiến so với kĩ thuật đệ qui thuần tuý, bất cứ khi nào chúng ta thấy lời giải cho cho bài toán đệ qui được lặp lại nhiều lần cho một dữ liệu vào, ta có thể cải tiến bằng qui hoạch động.

C.	Là chia bài toán lớn thành những bài toán con nhỏ hơn. Giải quyết bài toán con nhỏ, sau đó lưu kết quả với mục đích sử dụng lại để giải quyết bài toán tổng quát hơn.

D.	Cả A và B.

Đáp án : A,B,C. Vì đây là những ý tưởng của các nhóm 10 và 1

Câu 9 : Thuật toán là gì ?

A.	Thuật toán (giải thuật) là một tập hợp hữu hạn các hướng dẫn được xác định rõ ràng, có thể thực hiện được bằng máy tính, thường để giải quyết một lớp vấn đề hoặc để thực hiện một phép tính.

B.	Thuật toán (giải thuật) là một tập hợp các hướng dẫn được xác định rõ ràng, có thể thực hiện được bằng máy tính, thường để giải quyết một lớp vấn đề hoặc để thực hiện một phép tính.

C.	Thuật toán (giải thuật) là một tập hợp hữu hạn các hướng dẫn có thể thực hiện được bằng máy tính, thường để giải quyết một lớp vấn đề hoặc để thực hiện một phép tính.

D.	Thuật toán (giải thuật) là một tập hợp hữu hạn các hướng dẫn được xác định rõ rang thường để giải quyết một lớp vấn đề hoặc để thực hiện một phép tính.

Đáp án : A
